Ejercicio 2: Desarrolle una gráfica de linea usando el módulo pyplot de Matplotlib para producir la función:

$$ f(x) = x^2 |sen(x)| - 4 $$

Sobre el dominio $ 0 \leq x \leq 4 $. Asegurese de que los ejes y la curva esté correctamente etiquetadas. Agregue un titulo. 

```python
# Definir la función
def f(x):
    return x**2 * np.abs(np.sin(x)) - 4

# Crear valores de x en el intervalo [0, 4]
x_values = np.linspace(0, 4, 400)
# Calcular los valores de f(x)
y_values = f(x_values)

# Crear la gráfica de línea
plt.plot(x_values, y_values, label='$f(x) = x^2 |\sin(x)| - 4$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de $f(x) = x^2 |\sin(x)| - 4$')
plt.legend()
plt.grid(True)
plt.show()
```

Ejercicio 5:

Cree un gráfico en estructura de dos por dos usando **figure** para las siguientes funciones hiperbólicas en el dominio $ -1 \leq x \leq 1$

| Estructura del grafico | |
| ---------------------  | -- |
| $senh(x) \; vs \; x$ | $cosh(x) \; vs \; x$ |
| $tanh(x) \; vs \; x$ | $sech(x) \; vs \; x$ |

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de x
x_values = np.linspace(-1, 1, 100)

# Funciones hiperbólicas
senh_values = np.sinh(x_values)
cosh_values = np.cosh(x_values)
tanh_values = np.tanh(x_values)
sech_values = 1 / np.cosh(x_values)

# Crear la figura y los subgráficos
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Graficar senh(x) vs x
axs[0, 0].plot(x_values, senh_values, color='blue')
axs[0, 0].set_title('senh(x) vs x')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('senh(x)')

# Graficar cosh(x) vs x
axs[0, 1].plot(x_values, cosh_values, color='green')
axs[0, 1].set_title('cosh(x) vs x')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('cosh(x)')

# Graficar tanh(x) vs x
axs[1, 0].plot(x_values, tanh_values, color='red')
axs[1, 0].set_title('tanh(x) vs x')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('tanh(x)')

# Graficar sech(x) vs x
axs[1, 1].plot(x_values, sech_values, color='purple')
axs[1, 1].set_title('sech(x) vs x')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('sech(x)')

# Ajustar espacio entre subgráficos
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
